### IMPUTING MISSING VALUES IN ITEM_WEIGHT

```python
tot['Item_Weight']=tot['Item_Weight'].fillna(tot.groupby('Item_Identifier')['Item_Weight'].transform('mean'))
```

```python
tot.isnull().sum()
```

### IMPUTING THE MISSING VALUES OF OUTLET_SIZE WITH MODE

```python
from scipy.stats import mode
```

```python
outlet_size_mode = tot.pivot_table(values='Outlet_Size', columns='Outlet_Type',aggfunc=(lambda x:mode(x).mode[0]) )
outlet_size_mode
```

```python
miss_bool = tot['Outlet_Size'].isnull() 
sum(miss_bool)
```

```python
tot.loc[miss_bool,'Outlet_Size'] = tot.loc[miss_bool,'Outlet_Type'].apply(lambda x: outlet_size_mode[x])
```

```python
outlet_size_mode
```

```python
tot[(tot['Outlet_Type']=='Supermarket Type1')&(tot['Outlet_Size']=='High')]['Outlet_Size'].value_counts()
```

```python
tot.sample(10)
```

```python
tot.isnull().sum()
```

```python
tot['Item_Fat_Content']=tot['Item_Fat_Content'].replace({'LF':'Low Fat','reg':'Regular'})
tot['Item_Fat_Content'].value_counts()
```

```python
tot['Item_Fat_Content']=tot['Item_Fat_Content'].replace({'low fat':'Low Fat'})
tot['Item_Fat_Content'].value_counts()
```

```python
for i in tot.columns:
    print("{}\n".format(i),tot[i].value_counts())
```

```python
tot.info()
```

```python
tot['Outlet_Establishment_Year']=tot['Outlet_Establishment_Year'].astype('object')
```

```python
var_mod = ['Item_Fat_Content','Outlet_Location_Type','Outlet_Size','Item_Type_Combined','Outlet_Type','Outlet']
```

```python
data=pd.get_dummies(tot,columns=['Item_Fat_Content','Outlet_Establishment_Year','Outlet_Size','Outlet_Type',
                                'Outlet_Location_Type'])
```

```python
data.info()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
data['Item_Type']=le.fit_transform(data['Item_Type'])
```

```python
train=data[data['class']=='train']
```

```python
test=data[data['class']=='test']
```

```python
del train['class']
```

```python
X=num.drop(['Item_Outlet_Sales'],axis=1)
y=train['Item_Outlet_Sales']
```

```python
num1.shape
```

```python
Xtest=num1.drop(['Item_Outlet_Sales'],axis=1)
```

## LinearRegression

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
from sklearn.model_selection import train_test_split
Xtrain,X_test,ytrain,ytest=train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
lr.fit(Xtrain,ytrain)
ypred=lr.predict(X_test)
mse=mean_squared_error(ytest,ypred)
rmse=np.sqrt(mse)
print('MSE: ',mse)
print('RMSE: ',rmse)
```

```python
#Considering the whole train data  
```

```python
lr.fit(X,y)
```

```python
y_pred=lr.predict(Xtest)
```

```python
base1 = test[['Item_Identifier','Outlet_Identifier']]
```

```python
base1['Item_Outlet_Sales']=y_pred
```

```python
base1.to_csv('submission1.csv',index=True)
```

## Lasso and Ridge Regression:

```python
from sklearn.linear_model import Ridge,Lasso
from sklearn.model_selection import GridSearchCV
```

```python
lambdas=np.linspace(1,100,100)
params={'alpha':lambdas}
model=Ridge(fit_intercept=True)
grid_search=GridSearchCV(model,param_grid=params,cv=10,scoring='neg_mean_absolute_error')
```

```python
grid_search.fit(Xtrain,ytrain)
testpred=grid_search.predict(X_test)
mse=mean_squared_error(testpred,ytest)
rmse=((mse)**0.5)
print("MSE of RIDGE: ",mse)
print("RMSE of Ridge: ",rmse)
```

```python
grid_search.fit(X,y)
```

```python
grid_search.best_estimator_
```

```python
test_pred=grid_search.predict(Xtest)
```

```python
base1['Item_Outlet_Sales']=test_pred
```

```python
base1.to_csv('submission4.csv',index=False)
```
